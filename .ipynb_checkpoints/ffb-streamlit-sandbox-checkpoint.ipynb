```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
from datetime import date
```

```python
def extract_second_number(s):
    first_decimal_index = s.index('.')
    return s[first_decimal_index + 2:]
```

```python
def get_fpros_proj(position, base = 'standard'):
    print("Pulling: " + position + " - " + base + " values")
    
    if base == 'standard':
        url = "https://www.fantasypros.com/nfl/projections/" + position + ".php?week=draft"
    elif base == 'max':
        url = "https://www.fantasypros.com/nfl/projections/" + position + ".php?" + base +\
        "-yes=true&min-yes=&scoring=STD&week=draft"
    elif base == 'min':
        url = "https://www.fantasypros.com/nfl/projections/" + position + ".php?max-yes=&" + base +\
        "-yes=true&scoring=STD&week=draft"
        
        
        
        


    # Send a GET request to the URL
    response = requests.get(url)

    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.content, "html.parser")

        # Find the table with the desired class (you might need to inspect the page source to find the appropriate class)
        table = soup.find("table", class_="table")

        if table:
            # Extract table rows
            rows = table.find_all("tr")

            data = []  # To store table data

            for row in rows:
                # Extract table cells (columns) from each row
                cells = row.find_all("td")
                if cells:
                    row_data = [cell.get_text(strip=True) for cell in cells]
                    data.append(row_data)

            # Convert the data list into a pandas DataFrame
            df = pd.DataFrame(data[1:])  # Assuming the first row contains column headers

        else:
            print("Table not found on the page.")
    else:
        print("Failed to fetch the webpage.")


    if position == 'qb':
        col_names = ['player', 'pass_att', 'pass_cmp', 'pass_yds', 'pass_td', 'pass_int', \
                  'rush_att', 'rush_yds', 'rush_td', 'fumble_lost', 'fpts']
    elif position == 'rb':
        col_names = ['player','rush_att', 'rush_yds', 'rush_td', 'rec_att', 'rec_yds', 'rec_td','fumble_lost', 'fpts']
    elif  position == 'wr':
        col_names = ['player', 'rec_att', 'rec_yds', 'rec_td',\
                     'rush_att', 'rush_yds', 'rush_td', 'fumble_lost', 'fpts']
    elif position == 'te':
        col_names = ['player', 'rec_att', 'rec_yds', 'rec_td', 'fumble_lost', 'fpts']


    df.columns = col_names
    
    if base != 'standard':
        for col in df.columns:
            if col != 'player':
                df[col] = df[col].apply(extract_second_number)

                
    if base == 'max':
        df['player'] = df['player'].str.replace('high', '')
        df['player'] = df['player'].str.replace('low', '')

                    
    if base == 'min':
        df['player'] = df['player'].str.replace('high', '')
        df['player'] = df['player'].str.replace('low', '')

                
    team_abbreviations = ["BUF", 'MIA', 'NE', 'NYJ',
          "DEN", 'KC', 'LV', 'LAC',
          "BAL", 'CIN', 'CLE', 'PIT',
          "HOU", 'IND', 'JAC', 'TEN',
          "DAL", 'NYG', 'PHI', 'WAS',
          "ARI", 'LAR', 'SF', 'SEA',
          "CHI", 'DET', 'GB', 'MIN',
          "ATL", 'CAR', 'NO', 'TB']
    
    
    
    # Split player name and team abbreviation
    df["player"] = df["player"].str.strip()  # Remove any leading/trailing whitespace


    df["team"] = df["player"].str.extract(f'({"|".join(team_abbreviations)})$', expand=False)


    df["player"] =  df["player"].str[:-3] +\
    df["player"].str[-3:].replace(f'({"|".join(team_abbreviations)})$', '', regex=True).str.strip()



    df['position'] = position.upper()

    return(df)
```

```python
def calc_fpts(df_proj):


    pass_yds_pts = .04
    pass_td_pts = 4
    pass_int_pts = -1

    rush_yds_pts = .1

    rec_att_pts = 0
    rec_yds_pts = .1
    
    rush_red_td_pts = 6

    fumble_lost_pts = -2

    df_proj['fpts'] = df_proj.pass_yds*pass_yds_pts + df_proj.pass_td*pass_td_pts + df_proj.pass_int*pass_int_pts + \
                    df_proj.rec_att*rec_att_pts + df_proj.rec_yds*rec_yds_pts + df_proj.rec_td*rush_red_td_pts +\
                    df_proj.rush_yds*rush_yds_pts + df_proj.rush_td*rush_red_td_pts +\
                    df_proj.fumble_lost*fumble_lost_pts


    return(df_proj)

```

```python
def clean_fantasy_pull(dfs):
    

    # Concatenate all DataFrames into one
    df_proj = pd.concat(dfs, ignore_index=True)



    # Columns to convert to float
    float_columns = ['pass_att', 'pass_cmp', 'pass_yds', 'pass_td', 'pass_int',
                     'rush_att', 'rush_yds', 'rush_td', 'fumble_lost', 'fpts', 
                     'rec_att', 'rec_yds', 'rec_td']

    for col in float_columns:
        df_proj[col] = df_proj[col].str.replace(',', '').astype(float)

    df_proj.fillna(0, inplace=True)



    df_proj = calc_fpts(df_proj).sort_values(by='fpts', ascending = False)
    
    return(df_proj)
```

```python
#Core
positions = ["qb", "rb", "wr", "te"]
dfs = [get_fpros_proj(position) for position in positions]

df_proj = pd.concat(dfs, ignore_index=True)
df_proj['updated_date'] = (date.today()).strftime("%m.%d.%Y")
df_proj.to_csv('Data/fpros_projections.csv', index = False)


#Floor
positions = ["qb", "rb", "wr", "te"]
dfs = [get_fpros_proj(position, base = 'min') for position in positions]

df_proj = pd.concat(dfs, ignore_index=True)
df_proj.to_csv('Data/fpros_projections_floor.csv', index = False)


#Ceiling
positions = ["qb", "rb", "wr", "te"]
dfs = [get_fpros_proj(position, base = 'max') for position in positions]

df_proj = pd.concat(dfs, ignore_index=True)
df_proj.to_csv('Data/fpros_projections_ceil.csv', index = False)

```

```python
##Standard

positions = ["qb", "rb", "wr", "te"]
dfs = [get_fpros_proj(position) for position in positions]

df_proj = clean_fantasy_pull(dfs)


#High End

positions = ["qb", "rb", "wr", "te"]
dfs = [get_fpros_proj(position, base = 'max') for position in positions]

df_proj_ceil = clean_fantasy_pull(dfs)
df_proj_ceil['player'] = df_proj_ceil['player'].str.replace('high', '')
df_proj_ceil = df_proj_ceil[['player', 'fpts']]
df_proj_ceil.rename(columns={"fpts": "fpts_ceil"}, inplace=True)

#Low End


positions = ["qb", "rb", "wr", "te"]
dfs = [get_fpros_proj(position, base = 'min') for position in positions]

df_proj_floor = clean_fantasy_pull(dfs)
df_proj_floor['player'] = df_proj_floor['player'].str.replace('low', '')
df_proj_floor = df_proj_floor[['player', 'fpts']]
df_proj_floor.rename(columns={"fpts": "fpts_floor"}, inplace=True)


df_proj = pd.merge(df_proj, df_proj_ceil, how = 'left', on = 'player' )
df_proj = pd.merge(df_proj, df_proj_floor, how = 'left', on = 'player' )


df_proj = df_proj[['player', 'position', 'fpts', 'fpts_ceil', 'fpts_floor']]

df_proj['voltailty'] =  df_proj.fpts_ceil - df_proj.fpts_floor
df_proj['fpts_upside'] =  df_proj.fpts_ceil - df_proj.fpts
df_proj['fpts_downside'] =  df_proj.fpts_floor - df_proj.fpts


df_proj = df_proj.sort_values(by=['position', 'fpts'], ascending=[True, False])

# Assign ranks within each position group

df_proj['pos_rank'] = df_proj.groupby('position').cumcount() + 1
```

```python



#Change rank to pos rank
df_rb_proj = df_proj[(df_proj.position == 'rb') & (df_proj['pos_rank'] <= teams*(rb+flex))]
df_wr_proj = df_proj[(df_proj.position == 'wr') & (df_proj['pos_rank'] <= teams*(wr+flex))]
df_flex = df_rb_proj.append(df_wr_proj).sort_values(by=['fpts'], ascending=False)


flex_cutoff = df_flex.iloc[teams*(wr + rb + flex) - 1].fpts


qb_cutoff = df_proj[df_proj.position == 'qb'].iloc[teams*qb - 1].fpts
rb_cutoff = max(df_proj[df_proj.position == 'rb'].iloc[teams*(rb+flex) - 1].fpts,\
                min(df_proj[df_proj.position == 'rb'].iloc[teams*rb - 1].fpts, flex_cutoff))
wr_cutoff =  max(df_proj[df_proj.position == 'wr'].iloc[teams*(wr+flex) - 1].fpts,\
                min(df_proj[df_proj.position == 'wr'].iloc[teams*wr - 1].fpts, flex_cutoff))
te_cutoff = df_proj[df_proj.position == 'te'].iloc[teams*te - 1].fpts





print("QB Cutoff: " + str(qb_cutoff))
print("RB Cutoff: " + str(rb_cutoff))
print("WR Cutoff: " + str(wr_cutoff))
print("TE Cutoff: " + str(te_cutoff))

print("Flex Cutoff: " + str(flex_cutoff))

def assign_cutoff(row):
    if row['position'] == 'qb':
        return qb_cutoff
    elif row['position'] == 'rb':
        return rb_cutoff
    elif row['position'] == 'wr':
        return wr_cutoff
    elif row['position'] == 'te':
        return te_cutoff
    else:
        return None
    
df_proj['cutoff'] = df_proj.apply(assign_cutoff, axis=1)
df_proj['fpts_vbd'] = df_proj['fpts'] - df_proj.cutoff



#df_proj = df_proj.sort_values(by=['fpts_vbd'], ascending=False)

# Assign ranks within each position group

#df_proj['vbd_rank'] = df_proj.cumcount() + 1
```
